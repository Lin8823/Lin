# The New York Times News Title

```python
driverPath = 'chromedriver.exe'
browser = webdriver.Chrome(driverPath)
browser.implicitly_wait(1)   
browser.set_window_size(800, 700)  

url = 'https://www.nytimes.com/search?query=tesla'
browser.get(url)
title_list = []
desc_list = []
link_list = []
news_array = []
title = browser.find_elements_by_class_name('css-2fgx4k')
desc = browser.find_elements_by_class_name('css-16nhkrn')
link = browser.find_elements_by_class_name('css-1i8vfl5')
for tag in title:
    title_list.append(tag.text)
    print(tag.text)
for tag in desc:
    desc_list.append(tag.text)
    print(tag.text)
for tag in link:
    link_list.append(tag.get_attribute('href'))
    print(tag.get_attribute('href'))
news_array.append({"titles": title_list,
                            "urls": link_list,
                            "desc" :desc_list})
```
